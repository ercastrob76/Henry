## Selecciona los países donde se expandirán: Colombia, Argentina, Chile, México, Perú y Brasil.


```python
#Se crea una lista con los paises de expansion
Paises_exp=['Colombia','Argentina','Chile','Brazil','Mexico','Peru']
#Se usa .loc para filtrar los paises de interes
df_Paises_exp=df_latam.loc[df_latam['country_name'].isin(Paises_exp)]
df_Paises_exp
```

## Filtra los datos en fechas mayores a 2021-01-01.


```python
#Se filtran los datos por la fecha 2021-01-01
df_latam_desp_2021=df_Paises_exp[df_Paises_exp['date']>'2021-01-01']
```

## Compara a nivel de país para llenar valores faltantes.

### se filtra por la columna location_key con el fin de eliminar las subregiones para hacer el data frame mas ligero

```python
#Se seleccionan las columnas a analizar de acuerdo a previa evaluacion de correlacion, datos nulos y relevancia
df_latam_cols=df_latam_desp_2021.loc[:,['location_key' ,'date' ,'country_name' ,'new_confirmed' ,'new_deceased','cumulative_confirmed' ,'cumulative_deceased' ,'cumulative_vaccine_doses_administered' ,'population' ,'population_density','human_development_index' ,'population_age_30_39' ,'population_age_40_49' ,'population_age_50_59' ,'population_age_60_69','population_age_70_79' ,'population_age_80_and_older' ,'gdp_usd' ,'gdp_per_capita_usd' ,'infant_mortality_rate','nurses_per_1000' ,'physicians_per_1000' ,'average_temperature_celsius' ,'adult_male_mortality_rate' ,'adult_female_mortality_rate','pollution_mortality_rate' ,'comorbidity_mortality_rate' ]]

#se filtra por la columna location_key con el fin de eliminar las subregiones para hacer el data frame mas ligero
df_latam_paises = df_latam_cols[df_latam_cols['location_key'].isin(['AR','CL','CO','MX','PE','BR'])]
df_latam_paises.shape

#print(df_latam_cols.shape)
```

```python
#Se verifica la existencia de las columnas con valores nulos 
df_latam_null=df_latam_paises.isnull().sum()

print(df_latam_null)
```

## Realiza una limpieza preliminar de los datos, eliminando registros nulos y corrigiendo los tipos de datos donde sea necesario, trata con valores medios, valores anteriores o valores siguientes.

### Se recorre cada columna con valores nulos y estos son reemplazados por la media en cada pais

```python
#Se encuentran las columnas con vlaores nulos y se guardan en column_missing_values
columns_missing_values=df_latam_paises.columns[df_latam_paises.isnull().any()]
#Se recorre cada columna con valores nulos y estos son reemplazados por la media en cada pais
for column in columns_missing_values:
    df_latam_paises.loc[:,column]=df_latam_paises.groupby('country_name')[column].transform(lambda x: x.fillna(x.mean()))
```

## Guarda los datos filtrados en un archivo con el nombre DatosFinalesFiltrado.csv a fin de poder utilizarlo luego y no tener que repetir el proceso de filtrado y limpieza.


```python
#Se guarda el data frame como archivo .csv para respaldo en caso de necesitarlo
df_latam_paises.to_csv('C:\\Users\\Invite\\Downloads\\DatosFinalesFiltrado.csv')
```

## Crea una función que permita obtener la mediana, varianza y el rango. 

```python
def avg_rg_md(cols):        #Creacion de la funcion
    for i in cols.columns:
        if cols[i].dtype=='float64' or cols[i].dtype=='int64': #Seleccion de columnas de datos numericos
            print(i)
            print('Mediana:',cols[i].median())                 #Obtencion de la mediana
            print('varianza:',cols[i].var())                   #Obtencion de la varianza
            print('Rango', cols[i].max()-cols[i].min())        #Obtencion del rango
            print("Moda",cols[i].mode())                       #Obtencion de la moda
            print('---------------------------------')

avg_rg_md(df_latam_paises)
```

# Visualización de Datos con Matplotlib y Seaborn: Representa los hallazgos por cada país o de manera general a través de gráficos y/o visualizaciones. Debes incluir:


## Gráficos de barras de países contra el resto de variables.


```python
#Se seleccionan todas las columnas numericas que no se desean en los graficos para que no aparezcan
pop_age_cols=df_latam_paises.filter(like='population_age',axis=1).columns
#Se crea un nuevo df que excluya las columnas de edad
df_latam_clean2=df_latam_paises.drop(columns=list(pop_age_cols)) 
#El siguente ciclo for hace un barrido poor todas las columnas numericas y crea un grafico de barras de cada una por pais
for col in df_latam_clean2.select_dtypes(include='number').columns:
    plt.figure(figsize=(7,3))
    sns.barplot(data=df_latam_clean2, x='country_name', y=col, errorbar=None, hue='country_name')
    plt.xticks(rotation=45)
    plt.title(f'Comparacion de {col} por pais')
    plt.xlabel('Pais')
    plt.ylabel(f'{col}')
    plt.show()
```

## Mapas de calor para identificar correlaciones entre diferentes variables.


```python
#Se seleccionan las columnas de tipo numerico
df_latam_numbers=df_latam_paises.select_dtypes(include='number')
#Creacion de la matriz de correlacion
corr=df_latam_numbers.corr()
#Se crea una mascara que muestra solo la mitad de la matriz para no repetir valores
mask=np.triu(np.ones_like(corr,dtype=bool))
#Se crea una mascara para que solo se visualicen las correlaciones mayores a 0.5 en valor absoluto
corr=corr[corr.abs()>0.5]
#Creacion de la figura
plt.figure(figsize=(14,12))
#Creacion del mapa de calor
sns.heatmap(corr, mask=mask,cmap="coolwarm",annot=True, center=0, square=True, linewidths=0.5,cbar_kws={"shrink":0.8})
plt.title('Matriz de correlacion')
plt.xticks(rotation=90,ha='right')
plt.show()
```

## Histogramas gráficos de densidad y gráficos de densidad para entender la distribución de la incidencia de COVID-19 y las tasas de vacunación.


```python
#se seleccionan solo las columnas numericas que presentan cambios
mis_cols=df_latam_clean2.select_dtypes(include='number')
cols_variables=mis_cols.columns[mis_cols.nunique()>1]

# se realiza un barrido por las columnas con cambios y se grafica un histograma por cada una
for i, column in enumerate(cols_variables,1):
    plt.figure(figsize=(14,12))
    plt.subplot(len(cols_variables) // 3+1,3,i)
    sns.histplot(df_latam_clean2[column],bins=30, kde=True)
    plt.title(f'Histograma {column}')
    plt.xticks(rotation=45)
    plt.xlabel(column)
    plt.ylabel('Frecuencia')

    plt.tight_layout()

    plt.show()
```

# Avance 3

## Evolucion semanal de casos nuevos

```python
#Vamos a analizar como evolucionan los casos de COVID-19 en América Latina a lo largo del tiempo por semana

df_latam_paises_Fecha['week'] = df_latam_paises_Fecha.index.isocalendar().week
weekly_cases = df_latam_paises_Fecha.groupby(['country_name', 'week'])['new_confirmed'].sum().reset_index()
plt.figure(figsize=(14, 7))
for country in weekly_cases['country_name'].unique():
    country_weekly_cases = weekly_cases[weekly_cases['country_name'] == country]
    plt.plot(country_weekly_cases['week'], country_weekly_cases['new_confirmed'], label=country)

plt.xlabel('Semana del Año')
plt.ylabel('Casos Nuevos de COVID-19')
plt.title('Evolución Semanal de Casos Nuevos de COVID-19 en América Latina')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## Evolucion mensual de casos nuevos

```python
#Vamos a analizar como evolucionan los casos de COVID-19 en América Latina a lo largo del tiempo por mes
df_latam_paises_Fecha['month'] = df_latam_paises_Fecha.index.month
yearly_cases = df_latam_paises_Fecha.groupby(['country_name', 'month'])['new_confirmed'].sum().reset_index()
plt.figure(figsize=(14, 7))
for country in yearly_cases['country_name'].unique():
    country_yearly_cases = yearly_cases[yearly_cases['country_name'] == country]
    plt.plot(country_yearly_cases['month'], country_yearly_cases['new_confirmed'], label=country)

plt.xlabel('month')
plt.ylabel('Casos Nuevos de COVID-19')
plt.title('Evolución mensual de Casos Nuevos de COVID-19 en América Latina')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## Evolucion de nuevos casos y temperatura promedio

```python
# Se grafican simultaneamente los nuevos casos confirmados y la temperatura promedio en el tiempo para comparar su evolucion

plt.figure(figsize=(20, 7))
ax1 = plt.gca()
ax2 = ax1.twinx()
ax1.plot(df_latam_paises_Fecha.resample('ME').mean(numeric_only=True).index, df_latam_paises_Fecha.resample('ME').mean(numeric_only=True)['new_confirmed'], color='red', label='New Confirmed Cases')
ax2.plot(df_latam_paises_Fecha.resample('ME').mean(numeric_only=True).index, df_latam_paises_Fecha.resample('ME').mean(numeric_only=True)['average_temperature_celsius'], color='blue', label='Average Temperature', linestyle='--')

ax1.set_xlabel('Fecha')
ax1.set_ylabel('Nuevos casos confirmados', color='red')
ax2.set_ylabel('Temperatura promedio (°C)', color='blue')
plt.title('Nuevos casos de COVID-19 y temperatura promedio con el tiempo')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.show()
```

## Comparacion de estrategias de vacunacion 

```python
#Se realiza el cociente entre el acumulado de vacunas sobre la poblacion lo cual nos da el porcentajenumero de dosis de vacunados por persona por pais
vaccination_strategy = df_latam_paises_Fecha.groupby('country_name')['cumulative_vaccine_doses_administered'].max() / df_latam_paises_Fecha.groupby('country_name')['population'].mean()
vaccination_strategy.sort_values(ascending=False).plot(kind='bar', figsize=(12, 8), color='green')
plt.xlabel('País')
plt.ylabel('Dosis Administradas por cada persona')
plt.title('Comparación de Estrategias de Vacunación en América Latina')
plt.show()
```

## Comparacion de la tasa de mortalidad entre paises

```python
#Se obtiene el cociente entre los casos fatales sobre los casos confirmados con el fin de ver la tasa de mortalidad
df_latam_paises_Fecha['case_fatality_rate'] = df_latam_paises_Fecha['cumulative_deceased'] / df_latam_paises_Fecha['cumulative_confirmed']
fatality_rates = df_latam_paises_Fecha.groupby('country_name')['case_fatality_rate'].last().sort_values()

plt.figure(figsize=(10, 7))
fatality_rates.plot(kind='bar', color='red')
plt.xlabel('País')
plt.ylabel('Tasa de Mortalidad')
plt.title('Comparación de la Tasa de Mortalidad del COVID-19 Entre Países')
plt.show()
```

## Relacion tasa de vacunacion y nuevos casos

```python
# Como evoluciona la vacunación en América Latina y la reducción de casos de COVID-19
plt.figure(figsize=(14, 7))
ax1 = plt.gca()
ax2 = ax1.twinx()

#crea percentajes de vacunacion
df_latam_paises_Fecha['percentage_vaccinated'] = df_latam_paises_Fecha['cumulative_vaccine_doses_administered'] / df_latam_paises_Fecha['population'] * 100
df_latam_paises_Fecha.resample('ME').mean(numeric_only=True)['new_confirmed'].plot(ax=ax1, color='blue', label='Casos Nuevos')
df_latam_paises_Fecha.resample('ME').mean(numeric_only=True)['percentage_vaccinated'].plot(ax=ax2, color='green', label='Porcentaje de la Población Vacunada')


ax1.set_xlabel('Fecha')
ax1.set_ylabel('Casos Nuevos', color='blue')
ax2.set_ylabel('Porcentaje de la Población Vacunada', color='green')
ax1.set_title('Relación entre la Cobertura de Vacunación y la Reducción de Casos en América Latina')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.show()
```
