```python
# Asegurarse de que 'scheduled_departure' esté en formato datetime
flightdf['scheduled_departure'] = pd.to_datetime(flightdf['scheduled_departure'], format='ISO8601')

# Crear una función que modifique actual_departure dependiendo de scheduled_departure
def update_actual_departure(row):
    # Si 'actual_departure' está vacío o es '\N', asignar un valor aleatorio
    if pd.isna(row['actual_departure']) or row['actual_departure'] == "\\N":
        # Generar un retraso aleatorio entre 0 y 30 minutos
        delay_minutes = random.randint(0, 30)  # Este valor cambia aleatoriamente para cada fila
        actual_departure = row['scheduled_departure'] + timedelta(minutes=delay_minutes)
        return actual_departure
    return row['actual_departure']  # Si ya tiene un valor de actual_departure, no lo cambiamos



# Aplicar la función a cada fila del DataFrame
flightdf['actual_departure'] = flightdf.apply(update_actual_departure, axis=1)

flightdf['actual_departure'] = pd.to_datetime(flightdf['actual_departure'], format='ISO8601')

# Obtener la diferencia en minutos
flightdf['departure_delay_minutes'] = (flightdf['actual_departure'] - flightdf['scheduled_departure']).dt.total_seconds() / 60

# Imprimir el resultado para verificar
print(flightdf[['flight_id', 'scheduled_departure','scheduled_arrival', 'actual_departure','actual_arrival','departure_delay_minutes']])
```

```python

# Asegurarse de que 'scheduled_departure' esté en formato datetime
flightdf['scheduled_departure'] = pd.to_datetime(flightdf['scheduled_departure'], format='ISO8601')
# Asegurarse de que 'actual_departure' esté en formato datetime
flightdf['actual_departure'] = pd.to_datetime(flightdf['actual_departure'], format='ISO8601')
# Asegurarse de que 'scheduled_arrival' esté en formato datetime
flightdf['scheduled_arrival'] = pd.to_datetime(flightdf['scheduled_arrival'], format='ISO8601')


# Crear una función que modifique actual_arrival dependiendo de scheduled_arrival
def update_actual_arrival(row):
    # Si 'actual_departure' está vacío o es '\N', asignar un valor aleatorio
    if pd.isna(row['actual_arrival']) or row['actual_arrival'] == "\\N":
        # Generar un retraso aleatorio entre -10 y 17 minutos
        delay_minutes = random.randint(-10, 17)  # Este valor cambia aleatoriamente para cada fila, tambien se suma el retraso en la salida
        actual_arrival = row['scheduled_arrival'] + timedelta(minutes=delay_minutes)+timedelta(minutes=row['departure_delay_minutes'])
        return actual_arrival
    return row['actual_arrival']  # Si ya tiene un valor de actual_departure, no lo cambiamos

# Aplicar la función a cada fila del DataFrame
flightdf['actual_arrival'] = flightdf.apply(update_actual_arrival, axis=1)
# Asegurarse de que 'actual_departure' esté en formato datetime
flightdf['actual_arrival'] = pd.to_datetime(flightdf['actual_arrival'], format='ISO8601')

# Obtener la diferencia en minutos
flightdf['arrival_delay_minutes'] = (flightdf['actual_arrival'] - flightdf['scheduled_arrival']).dt.total_seconds() / 60

# Obtener la duracion del vuelo en minutos
flightdf['duration_mm']=(flightdf['actual_arrival']-flightdf['actual_departure']).dt.total_seconds()/60


# Calcular la diferencia entre actual_arrival y actual_departure
flightdf['duration'] = flightdf['actual_arrival'] - flightdf['actual_departure']

# Convertir la diferencia a formato HH:MM
flightdf['duration_hh_mm'] = flightdf['duration'].apply(lambda x: f"{x.seconds // 3600:02}:{(x.seconds % 3600) // 60:02}")


# Imprimir el resultado para verificar
print(flightdf[['flight_id', 'scheduled_departure','scheduled_arrival', 'actual_departure','actual_arrival','departure_delay_minutes','arrival_delay_minutes']])
```

```python
# Histograma de retrasos en la salida
plt.figure(figsize=(10, 3))
plt.hist(flightdf['departure_delay_minutes'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribución de retrasos en la salida (minutos)')
plt.xlabel('Retrasos (min)')
plt.ylabel('Frecuencia')
plt.show()

# Histograma de retrasos en la llegada
plt.figure(figsize=(10, 3))
plt.hist(flightdf['arrival_delay_minutes'], bins=30, color='orange', edgecolor='black')
plt.title('Distribución de retrasos en la llegada (minutos)')
plt.xlabel('Retrasos (min)')
plt.ylabel('Frecuencia')
plt.show()
```

Mayoría de los Vuelos Puntuales o con Retrasos Minimos: La barra más alta se encuentra en el rango de 0 minutos, lo que sugiere que una gran proporción de los vuelos salen a tiempo o con retrasos muy cortos.

Concentración de Retrasos en los Primeros Minutos: Se observa una disminución rápida en la frecuencia de retrasos a medida que aumenta el número de minutos. Esto indica que la mayoría de los retrasos se concentran en los primeros minutos y pocos vuelos experimentan retrasos extremadamente largos.

Mayoría de los Vuelos Llegan a Tiempo o con Pequeños Retrasos: Al igual que en el histograma de salidas, la barra más alta se encuentra en el rango de 0 minutos, lo que indica que una gran proporción de los vuelos llegan a tiempo o con retrasos mínimos.

Distribución Similar a la de Salidas: La forma general del histograma es similar al del histograma de salidas, con una disminución rápida en la frecuencia de retrasos a medida que aumenta el número de minutos. Esto sugiere que los factores que causan retrasos en las salidas también influyen en los retrasos en las llegadas.

```python
# Retrasos promedio por aeropuerto de salida
departure_delays = flightdf.groupby('departure_airport')['departure_delay_minutes'].mean().sort_values()

plt.figure(figsize=(16, 4))
departure_delays.plot(kind='bar', color='blue')
plt.title('Retrasos promedio por aeropuerto de salida')
plt.xlabel('Aeropuerto de salida')
plt.ylabel('Retraso promedio (min)')
plt.show()
```

```python
plt.figure(figsize=(10, 3))
sns.boxplot(x=flightdf['duration_mm'])
plt.title('Distribución de la duración de los vuelos (min)')
plt.xlabel('Duración (min)')
plt.show()
```

La mayoría de los vuelos tienen una duración similar: La caja es relativamente compacta, lo que sugiere que la mayoría de los vuelos tienen una duración similar.

Existen vuelos significativamente más largos: Los bigotes superiores y los outliers indican que hay una proporción de vuelos con una duración mucho mayor que el promedio.

```python
plt.figure(figsize=(10, 4))
plt.scatter(flightdf['departure_delay_minutes'], flightdf['arrival_delay_minutes'], alpha=0.5, color='blue')
plt.title('Relación entre retrasos de salida y llegada')
plt.xlabel('Retraso en salida (min)')
plt.ylabel('Retraso en llegada (min)')
plt.show()
```

Los retrasos en la salida son la principal causa de los retrasos en la llegada.

**Columnas Clave y de Identificación**

flight_id: Un identificador único para cada vuelo, como un número de vuelo.

flight_no: El número de vuelo asignado por la aerolínea.

**Información sobre el Vuelo**

scheduled_departure: La hora programada para el despegue del vuelo.

scheduled_arrival: La hora programada para la llegada del vuelo.

departure_airport: El código del aeropuerto de origen.

arrival_airport: El código del aeropuerto de destino.

status: El estado actual del vuelo (por ejemplo, Scheduled, Canceled, Delayed).

aircraft_code: El código del tipo de aeronave utilizada para el vuelo.

**Información Temporal**

actual_departure: La hora real a la que el vuelo despegó (si aplica).

actual_arrival: La hora real a la que el vuelo llegó (si aplica).
